```python
import pandas as pd

ventas_df = pd.read_csv('ventas.csv')
sucursales_df = pd.read_csv('sucursales.csv')
vehiculos_df = pd.read_csv('vehiculos.csv')
clientes_df = pd.read_csv('clientes.csv')
tiempo_df = pd.read_csv('tiempo.csv')
```

```python
ventas_df.drop_duplicates(inplace=True)

ventas_df.ffill(inplace=True)

if 'Fecha' in ventas_df.columns:
    ventas_df['Fecha'] = pd.to_datetime(ventas_df['Fecha'], format='%Y-%m-%d')

ventas_df = ventas_df.merge(sucursales_df, on='ID_Sucursal')
ventas_df = ventas_df.merge(vehiculos_df, on='ID_Vehículo')
ventas_df = ventas_df.merge(clientes_df, on='ID_Cliente')
ventas_df = ventas_df.merge(tiempo_df, on='ID_Tiempo')

ventas_df.to_csv('ventas_preprocesadas.csv', index=False)
```
